### #1.2

Crea una función similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una función para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numérica del dataframe
3. Una columna numérica del dataframe diferente a la anterior.
4. Un columna categórica (Opcional, valor por defecto None)
5. Un valor para el tamaño de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamaño
6. Un sexto argumento "escala", que permita manipular la escala del tamaño en el caso de que el quinto argumento sea una cadena. Este argumento multiplicará a los valores de la columna indicada en el parámetro anterior (así si quiero dividir por 10000, este sexto argumento tendrá que valer 1/10000)

La función debe pintar el diagrama de dispersión de la columna numérica 1, en el eje X, frente a la columna numérica 2, en el eje Y, usando un código de colores en función de la columna categórica y con el tamaño de los puntos según el argumento correspondiente. 

Si no se diera una columna categórica como argumento, el color no debe pasarse como parámetro a la función de Seaborn o Matplotlib que uses.

EXTRA: Añade un argumento que controle si la función pinta una leyenda para colores y tamaños o un colorbar y una leyenda para los tamaños. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesión en vivo)

```python
def scatter_plot(df, col_x, col_y, col_category=None, col_size=None, scale=1, show_legend=True):
    """
    Función para crear un diagrama de dispersión con parámetros flexibles.
    
    Argumentos:
    - df: DataFrame con los datos.
    - col_x: Nombre de la columna numérica para el eje X.
    - col_y: Nombre de la columna numérica para el eje Y.
    - col_category: Nombre de la columna categórica para el color (opcional).
    - col_size: Nombre de la columna o un valor fijo para los tamaños de los puntos.
    - scale: Factor de escala para los tamaños de los puntos.
    - show_legend: Booleano para mostrar/ocultar la leyenda.
    
    Salida:
    - Diagrama de dispersión generado.
    """
    plt.figure(figsize=(10, 6))
    
    # Calcular tamaños de los puntos (si se especifica una columna)
    if isinstance(col_size, str):
        sizes = df[col_size] * scale
    else:
        sizes = col_size if col_size else 50  # Tamaño predeterminado
    
    # Crear el gráfico según si hay o no columna categórica
    if col_category:
        scatter = sns.scatterplot(
            data=df,
            x=col_x,
            y=col_y,
            hue=col_category,
            size=sizes,
            legend=show_legend,
            sizes=(20, 200)
        )
    else:
        scatter = sns.scatterplot(
            data=df,
            x=col_x,
            y=col_y,
            size=sizes,
            legend=show_legend,
            sizes=(20, 200)
        )
    
    # Agregar etiquetas
    plt.xlabel(col_x)
    plt.ylabel(col_y)
    plt.title(f'Diagrama de dispersión: {col_x} vs {col_y}')
    
    # Mostrar u ocultar la leyenda solo si existe
    if not show_legend and scatter.legend_ is not None:
        scatter.legend_.remove()
    
    plt.show()
```

```python
# Ejemplo de llamada a la función
scatter_plot(
    df=df,
    col_x='elevation_m',  # Eje X
    col_y='population_total',  # Eje Y
    col_category='area_land_km2',  # Columna categórica (colores)
    col_size='population_total',  # Tamaño de los puntos basado en la población
    scale=0.0001,  # Escala para ajustar tamaños
    show_legend=True  # Mostrar la leyenda
)
```

## **Análisis de la gráfica: Tasa de Mortalidad en el Titanic**

- **No sobrevivieron (0):** La barra azul muestra a los que no sobrevivieron. **61.6%** de los pasajeros no sobrevivieron.

---

### **2. Interpretación bajo el contexto de un evento fatal**
El Titanic fue un desastre de magnitudes históricas:

- **Mayor mortalidad (61.6%):** 
  - La mayor parte de los pasajeros no sobrevivió, a pesar de ser considerado un barco "invencible". 

- **Menor tasa de supervivencia (38.4%):** 
  - Solo una pequeña parte de los pasajeros sobrevivió. Esto puede estar relacionado con la disponibilidad limitada de botes salvavidas, la desorganización durante el rescate y otros factores como la certeza equivocada de la seguridad del barco en general.

---

### **3. Reflexión sobre el desastre**
Este análisis pone de manifiesto la enorme **desigualdad en las probabilidades de supervivencia**. A pesar de las medidas de rescate que se tomaron, como el uso de los botes salvavidas, el **tamaño limitado de los botes** y la **desorganización en el proceso de evacuación**, fueron factores que contribuyeron a la alta mortalidad.

---

### **Conclusión**
La gráfica resalta la **mortalidad trágica** que ocurrió en el Titanic, donde la gran mayoría de los pasajeros no sobrevivieron, un hecho que subraya la gravedad y las lecciones que se aprendió del desastre.

```python
# Crear una columna para categorizar los grupos
def categorizar_pasajero(row):
    if row['age'] < 18 and row['sex'] == 'male':
        return 'Niños Hombres'
    elif row['age'] < 18 and row['sex'] == 'female':
        return 'Niñas'
    elif row['age'] >= 18 and row['sex'] == 'male':
        return 'Hombres Adultos'
    elif row['age'] >= 18 and row['sex'] == 'female':
        return 'Mujeres Adultas'
    else:
        return 'Sin Categoría'

df_titanic['Grupo'] = df_titanic.apply(categorizar_pasajero, axis=1)

tasa_supervivencia_por_grupo = df_titanic.groupby('Grupo')['survived'].mean() * 100

tasa_supervivencia_por_grupo = tasa_supervivencia_por_grupo[['Niñas', 'Niños Hombres', 'Mujeres Adultas', 'Hombres Adultos']]

plt.figure(figsize=(10, 6))
sns.heatmap(tasa_supervivencia_por_grupo.to_frame().T, annot=True, cmap='coolwarm', fmt='.1f', cbar=True)

plt.title('Tasa de Supervivencia por Grupo (Detallado)', fontsize=14)
plt.xlabel('Grupo', fontsize=12)
plt.ylabel('')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
