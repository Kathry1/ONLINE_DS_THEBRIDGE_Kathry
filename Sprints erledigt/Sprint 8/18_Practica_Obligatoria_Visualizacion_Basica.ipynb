### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# Ordenar por población total
df_sorted = df.sort_values(by="population_total", ascending=False).head(5)
```

```python
# Ciudades más pobladas
plt.bar(df_sorted["city"], df_sorted["population_total"], color=["blue", "gray", "olive", "brown", "pink"], alpha=0.7)
plt.title("Ciudades más pobladas de California", fontsize=14)
plt.ylabel("Población (miles)")
plt.xticks(rotation=45, ha='center')

# Agregar números encima de las barras
for i, value in enumerate(df_sorted["population_total"]):
    plt.text(i, value + 50, f"{value:.1f}", ha='center')

plt.show()
```

```python
# Histograma del área total ocupada
plt.hist(df["area_total_km2"], bins=50, color="steelblue", alpha=0.7, label="Área total")
plt.title("Distribución del área total ocupada (km²)", fontsize=14, fontname="Bradley Hand ITC")
plt.xlabel("Superficie (km²)")
plt.ylabel("Número de ciudades")
plt.legend()
plt.show()
```

```python
# Scatter del área de agua vs población
plt.scatter(df["area_water_km2"], df["population_total"], color="green", alpha=0.6)
plt.title("Relación entre el área de agua y la población", fontsize=14)
plt.xlabel("Área de agua (km²)")
plt.ylabel("Población (miles)")
plt.show()
```

```python
# Boxplot de la altura
sns.boxplot(data=df, x="elevation_m", color="red")
plt.title("Distribución de la altura de las ciudades", fontsize=14)
plt.xlabel("Altura (m)")
plt.show()
```

```python
# Crear un gráfico combinado
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Primer gráfico
axs[0, 0].bar(df_sorted["city"], df_sorted["population_total"], color=["blue", "gray", "olive", "brown", "pink"], alpha=0.7)
axs[0, 0].set_title("Ciudades más pobladas")
axs[0, 0].set_ylabel("Población (miles)")

# Segundo gráfico
axs[0, 1].hist(df["area_total_km2"], bins=50, color="steelblue", alpha=0.7)
axs[0, 1].set_title("Distribución del área total ocupada")

# Tercer gráfico
axs[1, 0].scatter(df["area_water_km2"], df["population_total"], color="green", alpha=0.6)
axs[1, 0].set_title("Relación área de agua y población")
axs[1, 0].set_xlabel("Área de agua (km²)")
axs[1, 0].set_ylabel("Población (miles)")

# Cuarto gráfico
sns.boxplot(data=df, x="elevation_m", color="red", ax=axs[1, 1])
axs[1, 1].set_title("Distribución de la altura")

plt.tight_layout()
plt.show()
```
